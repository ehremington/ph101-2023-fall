```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,20)
y = 2*x-3

with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.plot(x,y)
    plt.grid(True, lw=0.5, zorder=0)
    ax.set_xlabel('time (s)')
    ax.set_ylabel('velocity (m/s)')
    ax.axvline(0, color='k')
    ax.axhline(0, color='k')
```

```python
def vPlot(slope,yint):
    t = np.linspace(0,5,20)
    v = slope*x+yint
    fig = plt.figure(figsize=(3,3))
    ax = fig.add_subplot(aspect='equal', adjustable='datalim')
    ax.plot(t, v)
    plt.grid(True, 'both', lw=0.5, )
    ax.set_xlabel('time (s)')
    ax.set_ylabel('velocity (m/s)')
    ax.axvline(0, color='k')
    ax.axhline(0, color='k')
    fig.tight_layout()
    return fig
```

```python
fig1=vPlot(2,-3)
fig1.savefig('week4-v1-t.png',dpi=300, bbox_inches = "tight")
```

```python
def aPlot(value):
    t = np.linspace(0,5,20)
    a = value*np.ones(20)
    fig2 = plt.figure(figsize=(3,3))
    ax2 = fig2.add_subplot(aspect='equal', adjustable='datalim')
    ax2.plot(t, a)
    plt.grid(True, 'both', lw=0.5, )
    ax2.set_xlabel('time (s)')
    ax2.set_ylabel(r'acceleration (m/s$^2$)')
    ax2.axvline(0, color='k')
    ax2.axhline(0, color='k')
    fig2.tight_layout()
    return fig2
```

```python
fig2 = aPlot(-5)
fig2.savefig('week4-a1-t.png',dpi=300, bbox_inches = "tight")
```

```python
def rPlot(xi, vi, a):
    t = np.linspace(0,5,20)
    x = xi+vi*t+.5*a*t**2
    fig3 = plt.figure(figsize=(3,3))
    #ax3 = fig3.add_subplot(aspect='equal', adjustable='datalim')
    ax3 = fig3.add_subplot()
    ax3.plot(t,x)
    plt.grid(True, 'both', lw=0.5, )
    ax3.set_xlabel('time (s)')
    ax3.set_ylabel(r'position (m)')
    ax3.axvline(0, color='k')
    ax3.axhline(0, color='k')
    fig3.tight_layout()
    return fig3
```

```python
fig3 = rPlot(5,10,-3)
fig3.savefig('week4-x1-t.png',dpi=300, bbox_inches = "tight")
```
